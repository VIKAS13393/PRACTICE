```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
#creating empty list

author1=[]
time1=[]
title1=[]
desc1=[]
new_tube1=[]
```

```python
# activating the chrome browser

driver=webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver.exe")
driver.get("https://economictimes.indiatimes.com/archive.cms")a
```

```python
#extracting url of all the years

urls=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//a[@class='normtxt']")]
a=urls[182:194]       # scraping url of year 2015   
for i in a:
    driver.get(i)     #invoking url of each year
    fin=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//span[@class='normtxt']/div/table/tbody/tr/td/a")] # extracting url for each month of calendar
    for i in fin:
        driver.get(i)        # invoking each url of month
        try:
            new=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//ul[@class='content']/li/a")] # extracting the url of all the news
            for i in new:            
                driver.get(i)  # invoking each url of the news
                author=driver.find_element_by_xpath('//span[@class="ag"]')  # extracting name of the author of the news
                author1.append(author.text)
        except NoSuchElementException : # handling NoSuchElementException
                author1.append("---")
        try:
                time=driver.find_element_by_xpath('//div[@class="dtc vam artByline"]/time')   # extracting time of publish of news
                time1.append(time.text)
        except NoSuchElementException :
                time1.append("---")
        try:
                title=driver.find_element_by_xpath('//h1[@class="artTitle font_faus"]')  # extracting title of news
                title1.append(title.text)
        except NoSuchElementException :
                title1.append("---")  
        try:
                desc=driver.find_element_by_xpath('//div[@class="artText medium"]')   # extracting description of the news
                desc1.append(desc.text)
        except NoSuchElementException :
                desc1.append("---") 
        try :
                new_tupe=driver.find_element_by_xpath('//div[@class="clr breadCrumb contentwrapper"]/span[3]/a') # extracting genre of the news
                new_tube1.append(new_tupe.text)
        except NoSuchElementException :
                new_tube1.append("---") 
        
    
```

```python
 # creating the dataframe from the scraped data and taking only first 550 news
df=pd.DataFrame({"Date and time":time1[0:550],"Author":author1[0:550],"Vertical":new_tube1[0:550],
                "Headlines":title1[0:550],"Description":desc1[0:550]})
df
```

```python
df.to_csv("econoimcs_times.csv")     # store the scrapped news in the csv file
df.to_json("ecotimes.json")          # store the scrapped news in the json file
```
