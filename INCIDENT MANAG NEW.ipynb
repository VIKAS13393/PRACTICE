```python
columns_to_change_to_datetime = ['resolved_at', 'opened_at' ]
```

```python
for i in columns_to_change_to_datetime:
    df[i] = df[i].astype('datetime64[ns]')
```

```python
df['ticket_sol'] = ((df['resolved_at'] - df['opened_at']).dt.total_seconds())
```
