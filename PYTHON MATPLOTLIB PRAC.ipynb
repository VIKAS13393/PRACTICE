```python
Matplotlib is a plotting library for the python programming language and its numerical mathematics extension Numpy
```

```python
import matplotlib.pyplot as plt

%matplotlib inline  
 #  will prevent us to write plt.show() everytime
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
# plotting with matplotlib
#plt scatter

plt.scatter(x,y,c="g")  
plt.xlabel("X-axis")
plt.ylabel('Y-axis')
plt.title('Graph in 2D')
plt.savefig("new diagram.jpeg")  # create a jpeg/png file 
# we need not write plot.show() because of %matplotlin inline
```

```python
plt.plot(x,y,'r--',
        )
```

```python
# creating subplots

plt.subplot(2,2,1)  # 2,2,1(row,column,present plot)
plt.plot(x,y,'r*')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r*')
plt.subplot(2,2,2)
plt.plot(x,y,'g--')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r*')
plt.subplot(2,2,2)
plt.plot(x,y,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'b*')
plt.subplot(2,2,4)
plt.plot(x,y,'y--')
```

```python
# some complex X and Y parameters

x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.title("sine wave form")
plt.show()
```

```python
# bar plot

x=[2,8,20]
y=[11,16,9]

x2=[3,5,7]
y2=[6,17,8]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('bar graph')
plt.ylabel('Y-axis')
plt.xlabel('X-axis')

plt.show()
```

# PIE PLOT

```python
label='python','c++','ruby','java'
sizes=[215,130,234,118]
colors=['gold','lightskyblue','lightcoral','yellowgreen']
explode=(0.1,0.5,0,2.2)  # kitna crack hoga main circel se

plt.pie(sizes,explode=explode,label=label,color=colors)
autopct=
'%1.1f%%',shadow=True

plt.axis('equal')
plt.show()
    
```
