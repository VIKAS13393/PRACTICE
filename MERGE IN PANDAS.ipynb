```python
# IN CONCAT WE CAN ADD ONLY DATAFRAME completly without any conditions
# it will take all the columns 
```

```python
# Creating DataFrame
import pandas as pd
r1={
    'sub_id':['1','2','3','4'],
    'first':['class','bhora','debo','pinyu'],
    'second':['mishra','dad','bad','tat']
}
df1=pd.DataFrame(r1)
df1
```

```python
r2={
    'sub_id':['4','5','6','7','8'],
    'first':['vikas','moni','soni','naveen','sujit'],
    'second':['john','mathew','kumar','sah','duney']
}
df2=pd.DataFrame(r2)
df2
```

```python
r3={
    'sub_id':['1','2','3','4','5','6','7'],
     'test_id':['51','45','12','11','89','90','39']         
}

df3=pd.DataFrame(r3)
df3
              
```

```python
df_con=pd.concat([df2,df3],ignore_index=True)
df_con
```

```python
pd.merge(df2,df3)

# it will join with respect to common column
# and will take data of only common rows in that common column

```

```python
pd.merge(df1,df2)

# there is nothing common row among common column sub_id
```

```python
# when there is many common column,we can decide on which coluemn we wqnt to merge
pd.merge(df2,df3,on='sub_id')

# merge df2 and df3 with respect to sub_id( join all the column of df2,df3) butprint only common rows in sub_id
```

```python
pd.merge(df1,df2,on='first')
```

# MERGE WITH JOINS

```python
pd.merge(df2,df3,on='sub_id',how='inner')
```

```python
pd.merge(df2,df3,on='sub_id') # by default the join is inner join (AUB
```

```python
pd.merge(df2,df3,on='sub_id',how='outer') # by default the join is inner join (AUB
```

```python
df_con=pd.concat([df2,df3],ignore_index=True)
df_con
```

```python
df2.append(df3)   # just like concat
```

```python
df2.append(df3,ignore_index=True) # incegnore_true=true give new and proper seque
# just like concat 
 
```

# SET INDEX

```python
import pandas as pd

df=pd.DataFrame({"Day":[1,2,3,4],'visitors':[200,100,230,300],"rate":[20,30,40,50]})
print(df)
print("\n")
df1=df.set_index("Day")   # here day column become the index
print(df1)
```

```python
df=pd.DataFrame({"Day":[1,2,3,4],'visitors':[200,100,230,300],"rate":[20,30,40,50]})
df.set_index("Day",inplace=True)   # inplace = true means it will make change in same dataframe
print(df)
```
