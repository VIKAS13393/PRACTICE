# Pandas Visualization

Thses methods can be provided as the kind of argument to plot().
1. bar plot
2. hist plot
3. box plot
4. area plot
5. scatter plot

```python
# BAR PLOTrame

import pandas as pd
import numpy as np

df=pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
print(df)
df.plot.bar()
```

```python
import numpy as np

# get bar plot in the form of stack

df=pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
print(df)
df.plot.bar(stacked=True)
```

```python
# To get horizontal bar plots, use barh method:

df=pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
print(df)
df.plot.barh(stacked=True)
```

# Histograms can be plotted using plot.his() method

```python
df.plot.hist(bins=20)
```


BOX PLOT

```python
df=pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
df.plot.box()
```

AREA PLOT

```python
df=pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
df.plot.area()
```

```python
df=pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
df.plot.scatter(x='a',y='b')
```

```python
#This will concat 2 dataframe together
df_a=pd.DataFrame([['1','vikas','kumar'],['2','soni','kumari'],['3','moni','kumari']],columns=['id','first_name','Last_name'])
df_b=pd.DataFrame([['3','vikash','kumar'],['4','sonia','kumari'],['5','monia','kumari']],columns=['id','first_name','Last_name'])
df=pd.concat([df_a,df_b],ignore_index=True)  # here 
# by default  axis=0 that means join dataframe by rows
df
```

```python
#This will concat 2 dataframe together
df_a=pd.DataFrame([['1','vikas','kumar'],['2','soni','kumari'],['3','moni','kumari']],columns=['id','first_name','Last_name'])
df_b=pd.DataFrame([['3','vikash','kumar'],['4','sonia','kumari'],['5','monia','kumari']],columns=['id','first_name','Last_name'])
df=pd.concat([df_a,df_b],ignore_index=True,axis=1)  
# here axis=1 that means join dataframe by columns
df
```
