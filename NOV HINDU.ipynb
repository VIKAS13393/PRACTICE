```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
# creating empty list
time1=[]
title1=[]
desc1=[]
vertical=[]
author=[]

# activating the chrome browser
driver=webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver.exe")
driver.get("https://www.thehindu.com/archive/print/2018/11/")
```

```python
date=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//a[@class='ui-state-default']")]
for i in date[0:6]:
    driver.get(i)  
    link=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//ul[@class='archive-list']/li/a")] 
    for i in link:                 # extracting urls of first 6 news 
        driver.get(i)
        time.sleep(2)
        try:         # scraping title of news article          
            title=driver.find_elements_by_xpath("//h1[@class='title']")
            for i in title:
                title1.append(i.text)
        except NoSuchElementException :#handling no such element exception
            title1.append("---")
            
        try:        #scraping time of news article 
            till=driver.find_elements_by_xpath("//div[@class='ut-container']/span/none")
            for i in till:
                time1.append(i.text)
        except NoSuchElementException :#handling no such element exception
            time1.append("---") 
            
        try:      #scraping description of news article
            des=driver.find_element_by_xpath('//div[@class="article"]/div[3]/div[3]/p[1]')
            desc1.append(des.text)
        except NoSuchElementException :#handling no such element exception
            desc1.append("---")
            
        try:        # scraping author of news article
            auth=driver.find_elements_by_xpath("//span[@class='author-img-name 1']/a[2]")
            for i in auth:
                author.append(i.text)
        except NoSuchElementException :#handling no such element exception
            author.append("---")
            
        try:    #scraping vertical/top of news article
            ver=driver.find_elements_by_xpath("//a[@class='section-name']")
            for i in ver:
                vertical.append(i.text)
        except NoSuchElementException :#handling no such element exception
            vertical.append("---")      
```

```python
print(len(author))
```

```python
df1=pd.DataFrame({"DATE AND TIME":time1[0:3300],"AUTHOR":author[0:3300],"VERTICAL/TOP":vertical[0:3300],"HEADLINES":title1[0:3300],"DESCRIPTION":desc1[0:3300]})
df1
```
